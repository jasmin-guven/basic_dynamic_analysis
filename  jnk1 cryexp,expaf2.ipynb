```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "../Desktop/"
```

```python
jnk1_data = pd.read_csv(path + "jnk1_data.csv")
jnk1_ligand = jnk1_data["ligand"].to_numpy()
jnk1_af2 = jnk1_data["AF2_ddG"].to_numpy()
jnk1_exp = jnk1_data["exp_ddG"].to_numpy()
jnk1_cry = jnk1_data["crystal_ddG"].to_numpy()

```

```python
from sklearn.preprocessing import minmax_scale
err_scaled = minmax_scale(jnk1_data['error_IC50'].values)

# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# err_scaled =scaler.fit_transform(jnk1_data['error_IC50'].values.reshape(-2,2)).ravel()
```

```python
fig = plt.figure(figsize=(8, 8))
sns.set(style="ticks", font_scale=2.5)
ax = fig.add_subplot(111)
ax.set_xlabel(r"$\Delta \Delta G_{\mathrm{EXP}}$ (kcal mol$^{-1}}$)")
ax.set_ylabel(r"$\Delta \Delta G_{\mathrm{Crystal}}$ (kcal mol$^{-1}}$)")
ax.set_xlim(0,4)
ax.set_ylim(0,4)
x=[-0.5,4]
y=[-0.5,4]
ax.plot(x,y, linestyle ="--", c="grey")
ax.set_xticks(np.arange(-0.5,4))
ax.set_yticks(np.arange(-0.5,4))

# ax.set_xlim((0,4))
# ax.set_ylim((0,4))

fill_x = np.linspace(-1,4)

ax.fill_between(fill_x,fill_x-2,fill_x+2,color='gainsboro')
ax.fill_between(fill_x,fill_x-1,fill_x+1,color='darkgrey')

ax.errorbar(jnk1_exp[:-1],jnk1_cry[:-1],xerr=err_scaled[:-1],fmt='o',c='b')



plt.show()
plt.savefig(path+"thrombin_plot.jpg")
```
