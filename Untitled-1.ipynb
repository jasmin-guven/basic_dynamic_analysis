```python
import numpy as np
import MDAnalysis as mda
import MDAnalysis.analysis.rms
import MDAnalysis.analysis.distances
import matplotlib.pyplot as plt
```

```python
alphafold_data_path = "/home/yifan/Downloads/"
crystal_data_path ="/home/yifan/Desktop/"
ligand_names =['1a','1b','1c','1d','3a','3b','5','6a','6b','6e','7']
xtal_filenames = ['2ZFF__1__P_Dynamics', '2ZFF__1__P_Dynamics_3', '2ZFF__1__P_Dynamics_4', '2ZFF__1__P_Dynamics_5', '2ZFF__1__P_Dynamics_6', '2ZFF__1__P_Dynamics_7','2ZFF__1__P_Dynamics_8', '2ZFF__1__P_Dynamics_9', '2ZFF__1__P_Dynamics_10', '2ZFF__1__P_Dynamics_11','2ZFF__1__P_Dynamics_2']
af_filenames = ['AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics', 'AF2_dy_1', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_2', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_3', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_4', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_5', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_6', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_7', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_8', 'MODEL_AF-P00734-F1-model_v2_P_3u8t_362_9.model_Dynamics_9' ]
```

```python
for i in range(len(xtal_filenames)):
    print(xtal_filenames[i])
    print(af_filenames[i])
    af_ligand_frames = []
    with mda.lib.formats.libdcd.DCDFile(alphafold_data_path+f"{af_filenames[i]}.dcd") as trajectory_file:
        for frame in trajectory_file:
            af_ligand_frames.append(frame)

    xtal_ligand_frames = []
    with mda.lib.formats.libdcd.DCDFile(crystal_data_path+f"{xtal_filenames[i]}.dcd") as trajectory_file:
        for frame in trajectory_file:
            xtal_ligand_frames.append(frame)
    
    alphafold_ligand_first_frame = af_ligand_frames[0].xyz
    alphafold_ligand_universe = mda.Universe(alphafold_data_path+f"{af_filenames[i]}.prmtop", alphafold_data_path+f"{af_filenames[i]}.dcd")
    alphafold_ligand_reference_universe = mda.Universe(alphafold_data_path+f"{af_filenames[i]}.prmtop", alphafold_ligand_first_frame)

    xtal_ligand_first_frame = xtal_ligand_frames[0].xyz
    xtal_ligand_universe = mda.Universe(crystal_data_path+f"{xtal_filenames[i]}.prmtop", crystal_data_path+f"{xtal_filenames[i]}.dcd")
    xtal_ligand_reference_universe = mda.Universe(crystal_data_path+f"{xtal_filenames[i]}.prmtop", xtal_ligand_first_frame)

    xtal_ligand_structure = xtal_ligand_universe.select_atoms("resname MOL")
    xtal_reference_ligand_structure = xtal_ligand_reference_universe.select_atoms("resname MOL")

    af_ligand_structure = alphafold_ligand_universe.select_atoms("resname MOL")
    af_reference_ligand_structure = alphafold_ligand_reference_universe.select_atoms("resname MOL")

    xtal_rmsd_ligand = mda.analysis.rms.RMSD(xtal_ligand_structure, xtal_reference_ligand_structure)
    xtal_rmsd_ligand.run()

    af_rmsd_ligand = mda.analysis.rms.RMSD(af_ligand_structure, af_reference_ligand_structure)
    af_rmsd_ligand.run()


    xtal_ligand_rmsd_result = xtal_rmsd_ligand.results.rmsd.T
    print(xtal_ligand_rmsd_result)

    af_ligand_rmsd_result = af_rmsd_ligand.results.rmsd.T
    print(af_ligand_rmsd_result)

    af_ligand_rmsd_result = af_rmsd_ligand.results.rmsd.T
    print(af_ligand_rmsd_result)

    af_ligand_time = af_ligand_rmsd_result[1]
    af_ligand_rmsd_values = af_ligand_rmsd_result[2]

    xtal_ligand_time = xtal_ligand_rmsd_result[1]
    xtal_ligand_rmsd_values = xtal_ligand_rmsd_result[2]

    print(f"last crystal simulation time: {xtal_ligand_time[-1]}")
    print(f"last alphafold 2 simulation time: {af_ligand_time[-1]}")

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111)
    ax.plot(xtal_ligand_time, xtal_ligand_rmsd_values, color="pink", label="Crystal")
    ax.plot(af_ligand_time, af_ligand_rmsd_values, color="grey", label="AlphaFold 2")
    ax.set_xlabel("time (ps)", fontsize = 14)
    ax.set_ylabel(r"RMSD ($\AA$)", fontsize = 14)
    ax.legend(fontsize=14)
    ax.set_xlim(0, xtal_ligand_time[-1])
    fig.savefig(f"../basic_dynamic_analysis/plots/rmsd_ligand_comparison_{ligand_names[i]}.jpg")

    af_mean_rmsd_ligand = np.mean(af_ligand_rmsd_values)
    af_standard_deviation_rmsd_ligand = np.std(af_ligand_rmsd_values)

    xtal_mean_rmsd_ligand = np.mean(xtal_ligand_rmsd_values)
    xtal_standard_deviation_rmsd_ligand = np.std(xtal_ligand_rmsd_values)

    print(f"Average RMSD of ligand: ({af_mean_rmsd_ligand:.3f} \u00B1 {af_standard_deviation_rmsd_ligand:.3f}) \u212B")

    print(f"Average RMSD of ligand: ({xtal_mean_rmsd_ligand:.3f} \u00B1 {xtal_standard_deviation_rmsd_ligand:.3f}) \u212B")

    


```
