```python
for i in range(len(xtal_filenames)):
    print(xtal_filenames[i])
    print(af_filenames[i])
    af_ligand_frames = []
    with mda.lib.formats.libdcd.DCDFile(alphafold_data_path+f"{af_filenames[i]}.dcd") as trajectory_file:
        for frame in trajectory_file:
            af_ligand_frames.append(frame)

    xtal_ligand_frames = []
    with mda.lib.formats.libdcd.DCDFile(crystal_data_path+f"{xtal_filenames[i]}.dcd") as trajectory_file:
        for frame in trajectory_file:
            xtal_ligand_frames.append(frame)
    
    alphafold_ligand_first_frame = af_ligand_frames[0].xyz
    alphafold_ligand_universe = mda.Universe(alphafold_data_path+f"{af_filenames[i]}.prmtop", alphafold_data_path+f"{af_filenames[i]}.dcd")
    alphafold_ligand_reference_universe = mda.Universe(alphafold_data_path+f"{af_filenames[i]}.prmtop", alphafold_ligand_first_frame)

    xtal_ligand_first_frame = xtal_ligand_frames[0].xyz
    xtal_ligand_universe = mda.Universe(crystal_data_path+f"{xtal_filenames[i]}.prmtop", crystal_data_path+f"{xtal_filenames[i]}.dcd")
    xtal_ligand_reference_universe = mda.Universe(crystal_data_path+f"{xtal_filenames[i]}.prmtop", xtal_ligand_first_frame)

    xtal_ligand_structure = xtal_ligand_universe.select_atoms("resname MOL")
    xtal_reference_ligand_structure = xtal_ligand_reference_universe.select_atoms("resname MOL")

    af_ligand_structure = alphafold_ligand_universe.select_atoms("resname MOL")
    af_reference_ligand_structure = alphafold_ligand_reference_universe.select_atoms("resname MOL")
```
