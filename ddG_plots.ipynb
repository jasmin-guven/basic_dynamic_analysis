```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
jnk1_data = pd.read_csv("data/jnk1_data.csv")
jnk1_ligand = jnk1_data["ligand"].to_numpy()
jnk1_af2 = jnk1_data["AF2_ddG "].to_numpy()
jnk1_exp = jnk1_data["exp_ddG"].to_numpy()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(jnk1_exp, jnk1_af2, marker="o", c="pink")
ax.set_xlabel(r"$\Delta \Delta G_{\mathrm{EXP}}$ (kcal mol$^{-1}}$)", fontsize=14)
ax.set_ylabel(r"$\Delta \Delta G_{\mathrm{AF2}}$ (kcal mol$^{-1}}$)", fontsize=14)
ax.set_xlim(0,4)
ax.set_ylim(0,4)
```

```python
thrombin_data = pd.read_csv("data/thrombin_data.csv")
thrombin_ligand = thrombin_data["ligand"].to_numpy()
thrombin_af2 = thrombin_data["AF2_ddG"].to_numpy()
thrombin_exp = thrombin_data["exp_ddG"].to_numpy()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(thrombin_exp, thrombin_af2, marker="o", c="pink")
ax.set_xlabel(r"$\Delta \Delta G_{\mathrm{EXP}}$ (kcal mol$^{-1}}$)", fontsize=14)
ax.set_ylabel(r"$\Delta \Delta G_{\mathrm{AF2}}$ (kcal mol$^{-1}}$)", fontsize=14)
ax.set_xlim(0,2)
ax.set_ylim(0,2)
```
