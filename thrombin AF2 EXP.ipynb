```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path = "../Desktop/"
```

```python
thrombin_data = pd.read_csv(path + "thrombin_data.csv")
thrombin_ligand = thrombin_data["ligand"].to_numpy()
thrombin_af2 = thrombin_data["AF2_ddG"].to_numpy()
thrombin_exp = thrombin_data["exp_ddG"].to_numpy()
thrombin_cry = thrombin_data["crystal_ddG"].to_numpy()

```

```python
from sklearn.preprocessing import minmax_scale
err_scaled = minmax_scale(thrombin_data['error_dG'].values)

# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# err_scaled =scaler.fit_transform(thrombin_data['error_dG'].values.reshape(-1,1)).ravel()
```

```python
fig = plt.figure(figsize=(8, 8))
sns.set(style="ticks", font_scale=2.5)
ax = fig.add_subplot(111)
ax.set_xlabel(r"$\Delta \Delta G_{\mathrm{EXP}}$ (kcal mol$^{-1}}$)")
ax.set_ylabel(r"$\Delta \Delta G_{\mathrm{AF2}}$ (kcal mol$^{-1}}$)")
ax.set_xlim(-2,2.5)
ax.set_ylim(-2,2.5)
x=[-2,2.5]
y=[-2,2.5]
ax.plot(x,y, linestyle ="--", c="grey")
ax.set_xticks(np.arange(-2,2.5))
ax.set_yticks(np.arange(-2,2.5))

# ax.set_xlim((-2,3.5))
# ax.set_ylim((-2,3.5))

fill_x = np.linspace(-2,2.5)

ax.fill_between(fill_x,fill_x-2,fill_x+2,color='gainsboro')
ax.fill_between(fill_x,fill_x-1,fill_x+1,color='darkgrey')


ax.errorbar(thrombin_exp[:-2],thrombin_af2[:-2],xerr=err_scaled[:-2],fmt='o',c='b')


plt.show()
plt.savefig(path+"thrombin_plot.jpg")
```
